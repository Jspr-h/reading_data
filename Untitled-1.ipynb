```python

import pandas as pd  
import fastparquet as fp    
import matplotlib.pyplot as plt
```

```python
path = pd.read_parquet("C:/Users/Jesper/Downloads/archive/reading.parquet", engine="fastparquet")
path.to_csv("C:/Users/Jesper/Downloads/archive/reading.parquet.csv", index=False )
print(path)
```

```python
# Beskriver overordnede tendenser
path.describe()
```

```python

# Histogram over antallet af minutter læst, og antallet af gange det optræder
plt.figure(figsize=(10,6))
plt.hist(path['MINUTES_READING'], bins=30, color='skyblue', edgecolor='black')
plt.title('Distribution of Minutes Reading', fontsize=16)
plt.xlabel('Minutes Reading', fontsize=14)
plt.ylabel('Frequency', fontsize=14)   
plt.grid(axis='y', alpha=0.75)
plt.show()
```

```python
# definer en kategorisk rækkefølge rækkefølgen
categories = ['0-29 min', '30-59 min', '60-89 min', '90-119 min', '120+ min']

# opret variablen
path['READING_DURATION_CATEGORY'] = pd.Categorical(
    pd.cut(path['MINUTES_READING'],
           bins=[0, 29, 59, 89, 119, float('inf')],
           labels=categories,
           right=True),
    categories=categories,
    ordered=True  # Sæt til True for at bevare rækkefølgen
)


print(path['READING_DURATION_CATEGORY'].value_counts())  # Sorter efter kategorisk rækkefølge  
```

```python
# Histogram over antallet af minutter læst (i kategorier)
category_counts = path['READING_DURATION_CATEGORY'].value_counts().reindex(categories, fill_value=0)

plt.figure(figsize=(10,6))
category_counts.plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('Fordeling af læsetid (minutter)', fontsize=16)
plt.xlabel('Læsetid (minutter)', fontsize=14)
plt.ylabel('Hyppighed', fontsize=14)
plt.xticks(rotation=45)  # Drejer labels så det er pænere og mere læsbart
plt.grid(axis='y', alpha=0.75)
plt.show()
```

```python
# Vi vil gerne kigge på sammenhæng mellem reading_duration_category og reading_genre

#Vi laver en krydstabel
cross_tab = pd.crosstab(
    index=path["READING_GENRE"],
    columns=path["READING_DURATION_CATEGORY"],
    margins=True, 
    margins_name="Total"
)

print(cross_tab)
```

```python
# Vi laver et grupperet bar chart
cross_tab = pd.crosstab(path["READING_GENRE"], path["READING_DURATION_CATEGORY"])


# Plot
cross_tab.plot(kind="bar", stacked=True, figsize=(12, 6), colormap="viridis")
plt.title("Fordeling af læsetid pr. genre", fontsize=16)
plt.xlabel("Genre", fontsize=14)
plt.ylabel("Antal observationer", fontsize=14)
plt.xticks(rotation=45)
plt.legend(title="Læsetid")
plt.grid(axis="y", alpha=0.75)
plt.show()
```
